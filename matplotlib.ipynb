```python
import matplotlib.pyplot as plt


# initializing the data
x = [1, 2, 4, 9, 19, 25, 30]
y = [10, 25, 12, 48, 3, 15, 50]

a = [0, 10, 15, 20, 25]
b = [0, 20, 30, 15, 50]




# plotting the data

#subplot1
plt.subplot(2,1,1)
plt.grid(axis='y')
plt.plot(x, y, color='r',linestyle='dashed',linewidth=3,marker='.',markersize='15',mfc='black',mec='black')
#mec=markeredgecolor,mfc=markerfacecolor

#subplot2
plt.subplot(2,1,2)
plt.grid()
plt.plot(a, b, color='orange', marker='.',markersize='15',mfc='b',mec='b')



# Adding the title
plt.suptitle("Matplotlib Functions")


plt.show()
```

```python
#piechart
c = [10,30,50,10]
plt.subplot(2,2,3)
plt.pie(c,labels=['Electric','Pertol','Diesel','Hybrid'])
plt.show()
```

```python
#histogram
plt.hist(c)
plt.show() 
```
