```python
import pandas as pd
df = pd.read_csv('data.csv')
print(df.to_string()) 
```

```python
#removing none values
new_df = df.dropna()

print(new_df.to_string())
```

```python
#replacing none calories values with mean

df["Calories"].fillna(df["Calories"].mean(), inplace = True)
print(df)
```

```python
#correcting wrong date format

df['Date'] = pd.to_datetime(df['Date'])
print(df)
```

```python
#replacing wrong pulse values

a=df['Pulse'].mean()

for x in df.index:
    if df.loc[x, "Pulse"] > 300:
        df.loc[x, "Pulse"] = a
        
        
for x in df.index:
    if df.loc[x, "Pulse"] < 30:
        df.loc[x, "Pulse"] = a
        
print(df)
```
